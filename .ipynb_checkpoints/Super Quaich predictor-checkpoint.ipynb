# Predicting Super Quaich results from previous Scottish Cyclocross Series rounds

Import some libraries and define some helper functions for processing the data and parsing times

```python
import pandas as pd
import numpy as np
import glob
import datetime
import time
from sklearn import linear_model
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import train_test_split
from treeinterpreter import treeinterpreter as ti
import seaborn as sns
%matplotlib inline

def parse_time(raw_time):
    '''Parse the raw time strings into timedeltas for doing arithmetic on the times'''
    try:
        dt = datetime.datetime.strptime(str(raw_time), '%M:%S.%f')
        timedelta = datetime.timedelta(minutes=dt.minute, seconds=dt.second, microseconds=dt.microsecond)
        return timedelta
    except:
        return None


def process_data(df):

    #Remove non-rider rows
    df = df[(df['Name'] != 'Category Start') & (df['Name'] != 'Category Finish') & (df['Name'] != 'Timing Stopped')]
    #Remove rows without a lap time
    df = df[df['Lap Tm'].notnull()]

    # Convert lap times to timedeltas so that we can do arithmetic on them
    df['Lap Tm'] = df['Lap Tm'].apply(parse_time)

    #Hacky but effective way of getting median and mean for each rider. 
    #Inspiration from https://gist.github.com/tomfitzhenry/d36ebba697a1f3eeefcb
    df = df[['Name','Category','Lap Tm']].groupby(['Name','Category']).describe()

    #Munge the output to get only the bits we need (mean and median)
    df.index.names = ['Name', 'Category','Descriptor']
    df = df.unstack(level='Descriptor')
    df.columns = df.columns.droplevel(0)
    df = df[['50%','mean','count']]
    column_names = ['median', 'mean','lap count']
    df.columns = column_names

    #Make sure we don't get any oddities with people dropping out after one lap - often it's a short one
    df = df[df['lap count'] > 2]
    
    #Processing seems to re-format columns to strings so need to recast as timedeltas
    df['mean'] = pd.to_timedelta(df['mean'])
    df['median'] = pd.to_timedelta(df['median'])

    #Calculate difference from the fastest mean/median
    df['median_variance'] = df['median'] / df['median'].min()
    df['mean_variance'] = df['mean'] / df['mean'].min()

    df = df.sort_values('median_variance')
    return df

files = glob.glob('./scx2015files/*.csv')
```

Define the csv files that I want to look at and prepare the SCX data

```python
races = [{"Round": 1, "Index":[8,9,10]},
          {"Round": 2, "Index":[12,13,14]},
          {"Round": 3, "Index":[16,17,18]},
          {"Round": 6, "Index":[20,21,22]}]


all_df = pd.DataFrame()

for race in races:
    race_df = pd.DataFrame()
    
    for index in race["Index"]:
        raw_result = pd.read_csv(files[index])
        race_df = race_df.append(raw_result)
    
    race_df = process_data(race_df)
    race_df['Round'] = race['Round']
    
    all_df = all_df.append(race_df)
    
all_df.reset_index(inplace=True) 
all_df = pd.concat([all_df, pd.get_dummies(all_df['Category'])], axis=1)

df_grouped = all_df.groupby(['Name', 'Category','Junior Female', 'Junior Male', 'Senior',
       'Senior Female', 'Vet 40-49', 'Vet 50+', 'Vet Female'])['median_variance'].mean()

df_grouped = df_grouped.to_frame()
df_grouped.reset_index(inplace=True)
df_grouped = df_grouped.sort_values('median_variance')
df_grouped.reset_index(drop=True, inplace=True)
df_grouped.head(10)
```

Define the Super Quaich races from the files I've got and prepare the Super Q data

```python
superqindex = [{"Round":"SQ1", "Index":24},
               {"Round":"SQ2", "Index":26},
               {"Round":"SQ3", "Index":28},
               {"Round":"SQ4", "Index":30}]

sq_df1 = pd.DataFrame()

for race in superqindex:
    rawdf = pd.read_csv(files[race["Index"]])
    rawdf = process_data(rawdf)
    rawdf['Round'] = race["Round"]
    sq_df1 = sq_df1.append(rawdf)

sq_df1.reset_index(inplace=True)
sq_df = sq_df1[['Name','median_variance','Round']]
columns = ['Name','SQ_median_variance','SQ_Round']
sq_df.columns = columns

sq_grouped = sq_df.groupby(['Name'])['SQ_median_variance'].min()
sq_grouped = sq_grouped.to_frame()
sq_grouped.reset_index(inplace=True)
sq_grouped = sq_grouped.sort_values("SQ_median_variance")
sq_grouped.reset_index(drop=True, inplace=True)
sq_grouped.to_csv('sq_rank.csv')
sq_grouped.head(10)
```

Train a little linear regression model to predict Super Q results from the SCX results. Features are just the variance of the racer's median lap time from the fastest rider and dummy variables to define the category of the rider.

```python
joined = pd.merge(sq_grouped, df_grouped, on='Name')

train, test = train_test_split(joined, test_size = 0.2)
train_x = train[['Junior Female','Junior Male', 'Senior', 'Senior Female', 'Vet 40-49', 'Vet 50+','Vet Female', 'median_variance']]
train_y = train['SQ_median_variance']
test_x = test[['Junior Female','Junior Male', 'Senior', 'Senior Female', 'Vet 40-49', 'Vet 50+','Vet Female', 'median_variance']]
test_y = test['SQ_median_variance']

regr = linear_model.LinearRegression()
regr.fit(train_x, train_y)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print('Mean squared error: %.2f' % np.mean((regr.predict(test_x) - test_y) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(test_x, test_y))

lm = sns.lmplot('median_variance', 'SQ_median_variance', data=joined, fit_reg=False,hue="Category")
sns.plt.ylim(1,2)
sns.plt.xlim(1,2)
```

Now we can use the model to predict the Super Q performance on anyone that raced in the SCX series:

```python
predict_data = df_grouped[['Junior Female','Junior Male', 'Senior', 'Senior Female', 'Vet 40-49', 'Vet 50+','Vet Female', 'median_variance']]
prediction = pd.DataFrame(regr.predict(predict_data))
prediction.columns = ['Predicted']

joined_regr = df_grouped.join(prediction)
joined_regr = joined_regr.sort_values('Predicted')
joined_regr = joined_regr.merge(sq_grouped,how='left',on='Name')
joined_regr = joined_regr[['Name','Category','median_variance','Predicted','SQ_median_variance']]

column_names = ['Name','Category','SCX_median_variance','SQ_predicted_variance','SQ_median_variance']
joined_regr.columns = column_names
joined_regr = joined_regr[['Name','Category','SCX_median_variance','SQ_median_variance','SQ_predicted_variance']]

joined_regr['SCX_median_rank'] = joined_regr['SCX_median_variance'].rank()
joined_regr['SQ_predicted_rank'] = joined_regr['SQ_predicted_variance'].rank()

joined_regr.to_csv('regression_model_prediction.csv')
joined_regr.head(20)
```
